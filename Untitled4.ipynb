```python
import matplotlib.pyplot as plt
# data
x = [1,2,3,4,5,6,7,8]
y1 = [5,10,15,20,25,30,35,40] # Modified y1 to have 8 elements
y2 = [5,10,15,20,25, 28, 32, 35] # Modified y2 to have 8 elements
# area chart
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4 , label='Area 1-2')
plt.plot(x, y1, label='Line 1', marker='o') # Corrected 'labels' to 'label'
plt.plot(x, y2, label='Line 2', marker='o') # Corrected 'labels' to 'label'
#label and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('area chart')
# legend and display
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# Create a figure and a grid of subplots
fig, axs = plt.subplots(2, 2)  # 2x2 grid of subplots

# Plotting data on each subplot
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Plot 1')

axs[0, 1].plot(x, -y, 'tab:orange')
axs[0, 1].set_title('Plot 2')

axs[1, 0].plot(x, y**2, 'tab:green')
axs[1, 0].set_title('Plot 3')

axs[1, 1].plot(x, -y**2, 'tab:red')
axs[1, 1].set_title('Plot 4')

# Adding labels
for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Adjust layout
fig.tight_layout()

# Display the figure
plt.show()
```

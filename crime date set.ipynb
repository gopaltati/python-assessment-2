```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Date Rptd']
  ys = series['AREA']
  
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Date Rptd', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Vict Sex')):
  _plot_series(series, series_name, i)
  fig.legend(title='Vict Sex', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Date Rptd')
_ = plt.ylabel('AREA')
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Date Rptd']
  ys = series['Part 1-2']
  
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Date Rptd', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('Status Desc')):
  _plot_series(series, series_name, i)
  fig.legend(title='Status Desc', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Date Rptd')
_ = plt.ylabel('Part 1-2')
```
